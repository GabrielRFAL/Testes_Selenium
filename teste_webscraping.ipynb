```python
# Fazer o scrap de dados de uma página de livros específica para isso ("https://books.toscrape.com/")

# Importando as bibliotecas
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

# Inicializando uma instância do Chrome
# A classe Service é usada para iniciar uma instância do Chrome webdriver
servico = Service()

# O webdriver.ChromeOptions é usado para definir a preferência para o browser do Chrome
options = webdriver.ChromeOptions()

# Inicia-se a instância do Chrome WebDriver com as definidas 'options' e 'servico'
navegador = webdriver.Chrome(service=servico, options=options)

# Definição do URL a ser utilizado e abrindo o navegador
url = 'https://books.toscrape.com/'

navegador.get(url)

# Existem diversas formas de encontrar os elementos do HTML sendo elas:
# find_element(By.ID, "id")
# find_element(By.NAME, "name")
# find_element(By.XPATH, "xpath")
# find_element(By.LINK_TEXT, "link text")
# find_element(By.PARTIAL_LINK_TEXT, "partial link text")
# find_element(By.TAG_NAME, "tag name")
# find_element(By.CLASS_NAME, "class name")
# find_element(By.CSS_SELECTOR, "css selector")

# Nesse teste vamos capturar e colocar os seguintes dados em uma tabela: Nome e Quantidade em Estoque
# Utilizando o TAG_NAME e CLASS_NAME
# O Tag Name 'a' possui vários elementos, porem os títulos começam a ser exibidos no elemento 54
# por isso está sendo utilizado o [54] como na página existem 20 livros o elemento final é o 94 
# e os titulos são exibidos de 2 em 2 elementos portanto [54:94:2]
titleElements = navegador.find_elements(By.TAG_NAME, 'a')[54:94:2]

# Fazer com que para cada elemento busque o elemento 'title' para salvar em uma lista
titleList = [title.get_attribute('title') for title in titleElements]

# Agora queremos clicar no livro para checar quantos estão em estoque
titleElements[0].click()

# Para voltar a página - navegador.back()


# Agora para cada elemento na nossa lista iremos clicar nele encontrar quantos estão em estoque e colocá-lo em uma lista
# e repetir para todos
# o número de estoque está em uma classe (class="instock availability") como esta classe possui espaço utiliza-se somente a primeira palavra
# antes do espaço no caso em questão 'instock' para não dar erro
# Utilizar o find_element(By.CLASS_NAME, "class name")
# Utilizar o replace e transformar para inteiro para deixar somente a quantidade em estoque
int(navegador.find_element(By.CLASS_NAME, 'instock').text.replace("In stock (", "").replace(" available)", ""))

navegador.back()

# Agora basta colocar tudo o que foi feito em loop para ser realizado na lista toda

titleElements = navegador.find_elements(By.TAG_NAME, 'a')[54:94:2]
estoqueList = []

for title in titleElements:

    title.click()
    
    qtdEstoque = int(navegador.find_element(By.CLASS_NAME, 'instock').text.replace("In stock (", "").replace(" available)", ""))

    estoqueList.append(qtdEstoque)

    navegador.back()

# Colocar tudo isso dentro de um DataFrame utilizando o Pandas

dictDF = {'Titulo': titleList,
          'Estoque': estoqueList}

pd.DataFrame(dictDF)
```
